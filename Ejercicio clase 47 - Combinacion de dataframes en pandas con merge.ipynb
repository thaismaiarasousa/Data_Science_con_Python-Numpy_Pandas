```python
# Crear el df_peatones y un df adicinoal, df_pob, con la poblacion de cada uno de los distritos de Madrid.
```

```python
import pandas as pd
```

```python
df_peatones=pd.read_csv(r'C:\Users\thais\Desktop\Curso Python e Analise de Dados\Datasets+Data+Science+con+Python+-+Pandas\5.18\PEATONES_2020_mod.csv', delimiter=";",encoding="ISO-8859-1")
df_pob=pd.read_csv(r'C:\Users\thais\Desktop\Curso Python e Analise de Dados\Datasets+Data+Science+con+Python+-+Pandas\5.18\Distrito_poblacion.csv', delimiter=";",encoding="ISO-8859-1")
```

```python
# Realizar una combinacion de tipo inner de ambos dfs creando el df_combinado resultado de utilizar merge.
```

```python
df_combinado=df_peatones.merge(df_pob,on="DISTRITO", how="inner")
```

```python
df_combinado.head()
```

```python
# Cual distrito tiene mayor numero de peatones en el momento maximo de comparacion con su poblacion?
```

```python
# - crear un KPI = PEATONES / Poblacion_distrito
```

```python
df_combinado["KPI_peat_pob"]=df_combinado["PEATONES"]/df_combinado["Población_Distrito"]
```

```python
df_combinado.head()
```

```python
# Realizar un groupby para verificar el valor maximo del KPI en cada distrito
```

```python
df_combinado.groupby("DISTRITO")["KPI_peat_pob"].agg('max')
```
