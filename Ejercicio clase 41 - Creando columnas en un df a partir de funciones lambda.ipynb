```python
# Calcular a area da petala para todos os registros considerando cada petala como uma elipse - area= pi.altura.largura
```

```python
df_iris["area_petala"]=df_iris.apply(lambda x:x["longitud_pétalo"]*x["anchura_pétalo"]*np.pi,axis=1)
```

```python
# Qual a mediana da area da petala?
```

```python
df_iris["area_petala"].median()
```

```python
# Que valor de area eh superado somente pelos 10% das petalas?
```

```python
df_iris["area_petala"].quantile(0.9)
```
